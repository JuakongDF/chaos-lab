### Modelos SIN exponenciales (m√°s accesibles con √°lgebra b√°sica)

#### 1. **Modelo de Beverton-Holt** (Recursos Renovables)
$$N_{n+1} = \frac{R N_n}{1 + (R-1)N_n/K}$$
- **Aplicaciones:** Manejo de recursos naturales, forestales, reforestaci√≥n
- **Fen√≥meno:** Siempre converge a equilibrio estable (sin caos)
- **Datos:** Poblaciones de √°rboles, planes de reforestaci√≥n
- **Dificultad:** üü¢ F√°cil (solo divisi√≥n)

#### 2. **Modelo SIR Discreto** (Epidemiolog√≠a)
$$S_{n+1} = S_n - \beta S_n I_n$$
$$I_{n+1} = I_n + \beta S_n I_n - \gamma I_n$$
$$R_{n+1} = R_n + \gamma I_n$$
- **Aplicaciones:** Propagaci√≥n de enfermedades (gripe, COVID-19, sarampi√≥n)
- **Fen√≥meno:** Ondas epid√©micas, inmunidad de reba√±o
- **Datos:** OMS, MINSAL, CDC
- **Dificultad:** üü° Media (3 ecuaciones pero muy intuitivas)

#### 3. **Modelo de Hassell** (Competencia por Recursos)
$$N_{n+1} = \frac{\lambda N_n}{(1 + a N_n)^b}$$
- **Aplicaciones:** Insectos sociales, aves coloniales, lobos marinos
- **Fen√≥meno:** Efectos de agregaci√≥n espacial
- **Datos:** Poblaciones en colonias
- **Dificultad:** üü¢ F√°cil (una ecuaci√≥n con potencia)

#### 4. **Tent Map** (Caos Determinista Simple)
$$x_{n+1} = \begin{cases} 
\mu x_n & \text{si } x_n < 0.5 \\
\mu (1 - x_n) & \text{si } x_n \geq 0.5
\end{cases}$$
- **Aplicaciones:** Demostraci√≥n del efecto mariposa, criptograf√≠a
- **Fen√≥meno:** Caos puro con funci√≥n lineal
- **Dificultad:** üü¢ Muy f√°cil (solo if/else)

#### 5. **Lotka-Volterra Discreto** (Depredador-Presa)
$$H_{n+1} = H_n + r H_n (1 - H_n/K) - a H_n P_n$$
$$P_{n+1} = P_n + b a H_n P_n - m P_n$$
- **Aplicaciones:** Linces y liebres, zorros y conejos, conservaci√≥n
- **Fen√≥meno:** Oscilaciones poblacionales cl√°sicas
- **Datos:** Datos hist√≥ricos de linces/liebres en Canad√°
- **Dificultad:** üü° Media (2 ecuaciones pero muy visuales)

#### 6. **Modelo con Efecto Allee** (Extinci√≥n)
$$N_{n+1} = N_n + r N_n \left(1 - \frac{N_n}{K}\right) \left(\frac{N_n}{A} - 1\right)$$
- **Aplicaciones:** Especies en peligro (huemul, c√≥ndor, pud√∫)
- **Fen√≥meno:** Umbral de extinci√≥n
- **Datos:** Especies chilenas en peligro
- **Dificultad:** üü° Media (multiplicaciones, concepto importante)

### Modelos CON exponenciales (requieren m√°s matem√°tica)

‚ö†Ô∏è **Nota:** Estos modelos usan la funci√≥n exponencial $e^x$. Si no la has visto, consulta al profesor o elige uno de arriba.

#### 7. **Modelo de Ricker** (Pesca)
$$N_{n+1} = N_n e^{r(1 - N_n/K)}$$
- **Aplicaciones:** Gesti√≥n pesquera, sardinas, salm√≥n
- **Fen√≥meno:** Colapso por sobrepesca
- **Datos:** FAO, SUBPESCA, SERNAPESCA
- **Dificultad:** üü° Media (necesitas entender $e^x$)

#### 8. **Modelo de Nicholson-Bailey** (Control de Plagas)
$$H_{n+1} = \lambda H_n e^{-aP_n}, \quad P_{n+1} = c H_n (1 - e^{-aP_n})$$
- **Aplicaciones:** Control biol√≥gico de plagas agr√≠colas
- **Fen√≥meno:** Oscilaciones depredador-presa
- **Datos:** Plagas y parasitoides
- **Dificultad:** üü° Media (2 ecuaciones con exponenciales)

```python
# Instalar Plotly en JupyterLite
%pip install -q plotly
print("‚úì Plotly instalado correctamente")


import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
import IPython.display as display # <-- Nueva importaci√≥n para mostrar HTML
from plotly.offline import iplot # Mantener la importaci√≥n por si acaso


pio.templates.default = "plotly_white"
pio.renderers.default = "plotly_mimetype" 

print("‚úì Librer√≠as importadas correctamente")
print("‚úì Usando Plotly para gr√°ficos interactivos")


def tent_map(x, r):
    if x < 0.5:
        return r * x
    else:
        return r * (1 - x)


def iterate_map(model, x0, params, n_iterations):
    trajectory = np.zeros(n_iterations)
    trajectory[0] = x0
   
    for n in range(n_iterations - 1):
        trajectory[n + 1] = model(trajectory[n], *params)
   
    return trajectory


x0 = 0.1     
r = 1.95    
n_iterations = 100

trajectory = iterate_map(tent_map, x0, (r,), n_iterations)


df = pd.DataFrame({
    'Iteraci√≥n': np.arange(n_iterations),
    'Estado (x_n)': trajectory
})

fig = px.line(df, 
              x='Iteraci√≥n', 
              y='Estado (x_n)',
              title=f'Trayectoria Temporal del Tent Map (r={r}, x0={x0})')

fig.update_traces(mode='lines+markers', marker=dict(size=4))
fig.update_layout(xaxis_title='Iteraci√≥n (n)',
                  yaxis_title='$x_n$',
                  yaxis=dict(range=[0, 1]),
                  hovermode="x unified")


html_content = fig.to_html(include_plotlyjs='cdn')
display.display(display.HTML(html_content))
```

### 6.2 Diagrama de Bifurcaciones

*Explora c√≥mo cambia el comportamiento del sistema al variar un par√°metro*

```python

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
import IPython.display as display
from io import BytesIO

mpl.use('Agg')
import matplotlib.pyplot as plt 


def bifurcation_diagram(model, param_range, x0=0.1, n_skip=200, n_plot=100):
    """
    Genera diagrama de bifurcaciones
    
    Par√°metros:
    - model: funci√≥n del modelo (tent_map)
    - param_range: rango de par√°metros a explorar (r)
    - x0: condici√≥n inicial
    - n_skip: iteraciones a descartar (transitorio)
    - n_plot: iteraciones a graficar (atractor)
    """
    params = []
    values = []
    
    for r in param_range:
    
        x = x0
        for _ in range(n_skip):
            x = model(x, r)
        
 
        for _ in range(n_plot):
            x = model(x, r)

            if 0 <= x <= 1: 
                params.append(r)
                values.append(x)
    
    return np.array(params), np.array(values)


r_range = np.linspace(0.01, 2.0, 5000) # M√°s puntos para mayor detalle
n_skip_iterations = 200 # Descartar transitorio
n_plot_iterations = 100 # Graficar el atractor

r_vals, x_vals = bifurcation_diagram(tent_map, r_range, n_skip=n_skip_iterations, n_plot=n_plot_iterations)


fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(r_vals, x_vals, 
        ',k',  # ',k' significa: sin l√≠nea, marcadores muy peque√±os, color negro
        alpha=0.2) # Baja opacidad para revelar densidad de puntos

ax.set_title("Diagrama de Bifurcaciones del Tent Map")
ax.set_xlabel("Par√°metro de control r")
ax.set_ylabel("Valor asint√≥tico de $x_n$")
ax.set_ylim(0, 1) # El Tent Map siempre opera entre 0 y 1


buffer = BytesIO()
fig.savefig(buffer, format='png')
buffer.seek(0)

display.display(Image(data=buffer.read()))
plt.close(fig)

print("\n‚úì El Diagrama de Bifurcaciones para el Tent Map se ha generado y forzado a mostrarse como imagen PNG.")
```

### 6.3 Diagrama de Telara√±a (Cobweb) o Espacio de Fases

*Visualiza la din√°mica del sistema de forma geom√©trica*

**Para modelos de 1 variable (como el log√≠stico):** Usa diagrama de telara√±a (cobweb)

**Para modelos de 2+ variables (como SIR, Lotka-Volterra):** Usa diagrama de espacio de fases

Elige el que corresponda a tu modelo:

```python

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
from io import BytesIO

from IPython.display import Image, display 

mpl.use('Agg')
import matplotlib.pyplot as plt 


def cobweb_plot(model, r, x0, n_iterations=50):
    """
    Genera la figura del diagrama de telara√±a (cobweb plot)
    """
    # 1. Generar trayectoria
    x = np.zeros(n_iterations + 1)
    x[0] = x0
    for i in range(n_iterations):
        x[i + 1] = model(x[i], r)
    
    # 2. Preparar la curva de la funci√≥n f(x)
    x_range = np.linspace(0, 1, 1000)
    y_range = [model(xi, r) for xi in x_range]
    
    # 3. Preparar la trayectoria para la telara√±a (l√≠neas verticales y horizontales)
    cobweb_x = []
    cobweb_y = []
    for i in range(n_iterations):
        cobweb_x.extend([x[i], x[i], None])
        cobweb_y.extend([x[i], x[i+1], None])
        cobweb_x.extend([x[i], x[i+1], None])
        cobweb_y.extend([x[i+1], x[i+1], None])
    
    return x_range, y_range, cobweb_x, cobweb_y, x


r_values = [0.5, 1.0, 1.1, 1.95]
x0 = 0.1
n_iterations = 40


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 14))
plt.suptitle("Diagrama de Telara√±a (Cobweb Plot) del Tent Map", fontsize=18)

for i, r in enumerate(r_values):
    x_r, y_r, c_x, c_y, x_traj = cobweb_plot(tent_map, r, x0, n_iterations)
    
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    
    ax.plot(x_r, y_r, color='blue', linewidth=2, label=f'f(x), r={r}')
    ax.plot(x_r, x_r, color='black', linestyle='--', linewidth=1, label='y=x')
    ax.plot(c_x, c_y, color='red', linewidth=1, alpha=0.7)
    ax.plot(x_traj[0], x_traj[0], 'o', color='green', markersize=8, label=f'Inicio $x_0={x0}$')
    
    # Configuraci√≥n de ejes
    ax.set_title(f'$r = {r}$: {["Extinci√≥n", "L√≠mite", "Ciclos", "Caos Cercano"][i]}', fontsize=14)
    ax.set_xlabel('$x_n$')
    ax.set_ylabel('$x_{n+1}$')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.legend(loc='upper right', fontsize=8)


plt.tight_layout(rect=[0, 0, 1, 0.96])

buffer = BytesIO()
fig.savefig(buffer, format='png')
buffer.seek(0)

display(Image(data=buffer.read())) 
plt.close(fig)

print("\n‚úì Los Diagramas de Telara√±a se han generado y forzado a mostrarse como imagen PNG.")
```

### 6.5 An√°lisis Cuantitativo

*Calcula m√©tricas relevantes*

```python

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
from io import BytesIO
from IPython.display import Image, display


mpl.use('Agg')
import matplotlib.pyplot as plt 


# El caos ocurre en r=2, y el sistema diverge despu√©s.
r_test = np.linspace(0.01, 2.0, 500) 


# Lambda (Œª) = ln(r)
lyap_vals = np.log(r_test)


fig, ax = plt.subplots(figsize=(10, 6))

# Exponente de Lyapunov
ax.plot(r_test,
        lyap_vals,
        color='blue', 
        linewidth=2, 
        label='Exponente de Lyapunov (Œª = ln(r))')

# L√≠nea en y=0 (frontera estable/ca√≥tico)
ax.axhline(
    y=0, 
    color='red', 
    linestyle='--', 
    linewidth=1,
    label="Frontera Estable/Ca√≥tico (Œª=0)"
)

# Sombreado de regiones
ax.fill_between(r_test, lyap_vals, 0, where=(lyap_vals < 0), color='green', alpha=0.1, label='R√©gimen Estable (Œª < 0)')
ax.fill_between(r_test, lyap_vals, 0, where=(lyap_vals > 0), color='red', alpha=0.1, label='R√©gimen Ca√≥tico (Œª > 0)')


# Configuraci√≥n del gr√°fico
ax.set_title("Exponente de Lyapunov (Œª) vs. Par√°metro r para el Tent Map")
ax.set_xlabel("Par√°metro de control r")
ax.set_ylabel("Exponente de Lyapunov (Œª)")
ax.set_xlim(0, 2)
ax.set_ylim(-3, np.log(2) + 0.1) 
ax.legend()
ax.grid(True, linestyle=':', alpha=0.7)


# El caos ocurre cuando Œª > 0. Esto sucede cuando ln(r) > 0, o r > 1.
r_caos = 1.0

# Imprimir Resultados
print(f"\n--- An√°lisis Cuantitativo del Tent Map ---")
print(f"La f√≥rmula exacta es: Œª = ln(r)")
print(f"R√©gimen Estable (Œª < 0): 0 < r < {r_caos:.2f}")
print(f"R√©gimen Ca√≥tico (Œª > 0): {r_caos:.2f} < r ‚â§ 2.0")
print(f"Valor m√°ximo de Caos (en r=2): Œª_max = ln(2) ‚âà {np.log(2):.3f}")


buffer = BytesIO()
fig.savefig(buffer, format='png')
buffer.seek(0)

display(Image(data=buffer.read()))
plt.close(fig)
```
