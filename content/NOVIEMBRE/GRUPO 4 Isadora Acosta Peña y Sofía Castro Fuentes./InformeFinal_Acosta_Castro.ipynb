Abstract (Resumen)

En este trabajo se investiga el comportamiento de las poblaciones del Pejerrey chileno (*Basilichthys australis*) y la Carpa común europea (*Cyprinus carpio carpio*) (una especie introducida) en las aguas chilenas y la interacción entre ellas, verificando si es sostenible a lo largo del tiempo dependiendo de su población inicial, en este caso en 200 temporadas reproductivas. Se utilizó el llamado modelo de Hassell para la competencia por recursos, siendo adaptado para incluir la competencia tanto intra- como inter-especie en su fórmula, siendo N1, t+1= (R1 - N1,t)/ (1 + a11N1,t + a12N2,t)b1, para diferentes valores de N. Se encontró que la carpa, al tener una mayor capacidad reproductiva en comparación al pejerrey, acaba por dominarlo y llevar al pejerrey a un porcentaje mínimo de población o a una extinción, ya sea en escenarios con ventaja (valores de N_c > 0.5), desventaja (0.2 < N_c < 0.5) o equilibrio (N_c = 0.5). Los resultados indican que medidas de control a la especie invasora son de gran importancia para prevenir la gran disminución o extinción del pejerrey chileno, comprobando además que a mayor tasa de reproducción y eficiencia de recolección de recursos, las carpas siempre terminan dominando a los pejerreyes con mínimas diferencias entre variaciones dada su naturaleza.

En sistemas acuáticos donde ambas especies coexisten, se ha observado que la carpa puede modificar profundamente el ecosistema que habita, esto al aumentar la turbidez y consumir grandes cantidades de zooplancton, insectos y pequeños invertebrados. Estos mismos recursos constituyen la base alimentaria del pejerrey (el cual posee una capacidad reproductiva y supervivencia inmensamente menor en comparación), lo que genera un escenario de competencia intraespecífica desigual, el cual se traduce en dinámicas poblacionales contrastantes y posibles desplazamientos del pez nativo. En este contexto, los modelos matemáticos, como el modelo de Hassel, son los más adecuados para representar la competencia por recursos entre miembros de una misma especie o de varias al ser adaptada la fórmula, como la carpa y el pejerrey chileno que es el caso que estudiamos, al ser un modelo capaz de predecir comportamientos poblacionales en base a las tasas de reproducción, la supervivencia e intensidad de competencia, simulando escenarios de coexistencia o exclusión competitiva y ayudando en la elaboración de estrategias de cuidado y conservación de la especie.

En consecuencia, se plantea la **hipótesis** de que la carpa, al ser más eficiente en el uso de sus recursos y tener una mayor capacidad reproductiva en comparación al pejerrey, acabará por dominarlo y llevar al pejerrey a un porcentaje mínimo de población o a una extinción, ya sea en escenarios con ventaja (valores de N_c > 0.5), desventaja (0.2 < N_c < 0.5) o equilibrio (N_c = 0.5).


##*Sección 3: Marco teórico*

El modelo matemático a utilizar es el **Modelo de Hassell** para competencia por recursos, habitualmente modelada para la competencia intraespecie mediante:

$$N_{n+1} = \frac{\lambda N_n}{(1 + a N_n)^b}$$



Pero modificada para incluir tanto la competencia intra- como inter-especie de manera:

$$N_{1, n+1} = \frac{R_1 N_{1,t}}{(1 + a_{11} N_1,t + a_{12} N_2,t)^b}$$

$$N_{2, n+1} = \frac{R_2 N_{2,t}}{(1 + a_{22} N_1,t + a_{21} N_1,t)^b}$$

Donde:

- $N_{n}$ = Tamaño poblacional de la especie en el tiempo $t$.
- $λ$ o $R$ = Tasa máxima de crecimiento (sin competencia).
- $a$ = Intensidad de competencia intra e inter-específica.
- $b$ = Parámetro que controla la fuerza de la competencia y qué tan rápido esta reduce el crecimiento cuando aumenta la densidad.
- $_1$ = Especie 1.
- $_2$ = Especie 2.

###*Significado de los parámetros*

- $N_{n}$: Representa la proporción entre el tamaño poblacional de las dos especies, con valores de $R_c < 0.5 < R_p$ o viceversa muestran una ventaja de una especie sobre la otra y R_c = 0.5 = R_p evidenciando un empate de condiciones iniciales.

- $R$: En este caso el tamaño poblacional está dado por la tasa de reproducción por desove, cantidad de desoves al año y la tasa de supervivencia de las especies.

- $a$: Se representa la intensidad de competencia entre la misma especie (11/pp y 22/cc) y entre las dos especies (12/pc para el efecto del pejerrey sobre la carpa y 21/cp para el efecto de la carpa sobre el pejerrey), donde al ser $a$ > 5 el efecto es alto y $a$ < 5 presenta un bajo efecto sobre la otra especie.

- $_1$: En este caso el pejerrey [p].

- $_2$: En este caso la carpa [c].

###*Comportamiento esperado*

- Cuando la densidad está controlada por competencia y el crecimiento poblacional se modera, la población se acerca a un valor estable, comportándose como un sistema autorregulado.

- En caso de que la competencia sea muy fuerte o R sea demasiado bajo, se predice un a falta de recursos y mortalidad alta.

- En caso de haber 9una diferencia significativa de R entre ambas especies ($R_1 < R_2$), la de R mayor con competencia moderada dominará sobre la otra, llevándola a una extinción.

- Cuando una especie se limita a si misma más que a la otra ($a_11 > a_12 y a_22 > a_21$), se llega a una coexistencia estable.

## *Sección 5: Código*

```python
# Instalar Plotly en JupyterLite
%pip install -q plotly
print("✓ Plotly instalado correctamente")
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Configuración de Plotly
import plotly.io as pio
pio.templates.default = "plotly_white"

print("✓ Librerías importadas correctamente")
print("✓ Usando Plotly para gráficos interactivos")
```

```python
import matplotlib.pyplot as plt

# Modelo de Hassell

# Parámetros

#Tasa reproductiva (R)
R_p = 30720 #pejerrey
R_c = 240000 #carpa

maxR=30000*7

R_c=R_c/maxR
R_p=R_p/maxR

#Intensidad de competencia (a)
a_pp = 1.0   #Competencia intra-específica
a_cc = 1.0

a_cp = 0.128 #Competencia inter-específica
a_pc = 7.8

#Parámetro b (curvatura del efecto de competencia)
b_p = 1.3
b_c = 1.0

#Simulación

steps = 200  #En ciclos reproductivos
N_p = np.zeros(steps)
N_c = np.zeros(steps)

N_p[0] = 0.5
N_c[0] = 0.5

#Aplicación del modelo de Hassell (para ambas especies)

for t in range(steps - 1):
    N_p[t + 1] = (R_p * N_p[t]) / ((1 + a_pp*N_p[t] + a_pc*N_c[t])**b_p) #Modelo para pejerrey en relación a la carpa
    N_c[t + 1] = (R_c * N_c[t]) / ((1 + a_cc*N_c[t] + a_cp*N_p[t])**b_c)  #Modelo para carpa en relación al pejerrey

#Graficar

plt.figure(figsize=(10, 6))
plt.plot(N_p, label="Pejerrey chileno")
plt.plot(N_c, label="Carpa común")
plt.xlabel("Tiempo (ciclos reproductivos)")
plt.ylabel("Población")
plt.title(f"Modelo de Hassell")
plt.legend()
plt.grid(True)
plt.show()
```

##*Sección 6: Análisis*

Esto indica que, bajo las condiciones establecidas, la competencia intraespecífica (a_pp) regula su crecimiento pero la competencia interespecífica la domina. Mientras que la carpa, independientemente de su N, se estabiliza en valores entre el 20% y el 10% a tras un período más largo de oscilaciones.

**Confirma esto entonces la hipótesis** planteada: sin importar el punto inicial de las poblaciones de ambas especies, la eficiencia de uso de recursos y la capacidad de reproducción de la carpa sobrepasan con creces las del pejerrey, llevando a esta especie a su rápida desaparición.

En cuanto a la **literatura** citada, los trabajos de Hassel y Comins muestran el funcionamiento del modelo discreto, mostrando que incluso iteraciones simples pueden tener una alta sensibilidad al cambio en los parámetros establecidos, evidenciado en el comportamiento en las primeras fases de iteraciones en el gráfico. Por otro lado, trabajos como el de Anazawa muestran como la desigualdad entre recursos y competencia en las especies pueden amplificarse ante diferencias en las condiciones iniciales, confirmando nuevamente que sistemas ecológicos como el presentado pueden mostrar reusltados altamente dependientes del estado inicial.

**En lo práctico**, esto se traduce en que las carpas introducidas significan una gran amenaza para la supervivencia los pejerreyes nativos de la zona, y probablemente otras especies nativas de similares características. Aquello hace necesarias medidas de control de la especie invasora, limitando su número y entorno con el cual interactúa para que no influyan en el medioambiente de su competencia.

Por otro lado, el sistema presenta ciertas **limitaciones** más allá de la falta de información del caso en específico, como la no consideración de la depredación de huevos y larvas de pejerrey por parte de las mismas carpas, las etapas y madurez sexual de la población antes de poder reproducirse, la disponibiloidad variable de los recursos según la estación del año, la competencia con otras especies dentro del mismo hábitat y, claro está, el azar.
