Polilla de la Vid: Reproducción
Como grupo, decidimos trabajar la especie de la polilla de la vid,una plaga muy conocida en la zona
central de Chile y en los viñedos de nuestro país.
Trabajamos específicamente plaguicidas para su control, construimos nuestro gráfico y su
crecimiento poblacional en base al plaguicida utilizado, comparando los resultados.
Nos motivó mucho trabajar en esta especie, pues la producción de uvas, es una parte muy
importante de la economía chilena, ya que de ahí nacen nuestros vinos reconocidos mundialmente.
Para ayudar en el control de la especie intentamos predecir el comportamiento de esta una vez se
haya aplicado el plaguicida. Para ello, decidimos trabajar el mapa logístico,
Nuestra pregunta de Investigación fué: ¿Cómo se ve afectado el crecimiento poblacional de la polilla
de la vid tras la aplicación de distintos plaguicidas?
Nuestra hipótesis es que depende de la eficiencia del plaguicida.
No serviría de nada aplicar un plaguicida de baja eficiencia, pues no reduciría tanto la tasa de
reproducción de la polilla.

Análisis del Gráfico
Con una población inicial de 100 polillas los plaguicidas tuvieron muy distintos resultados.
Sin plaguicidas la plaga se disparó y creció rápidamente.
Después, el plaguicida A resultó horripilantemente, pues eliminó exclusivamente al 50% de la
población, que inicialmente fué impactante en el crecimiento de la especie, pero que después fué
insignificante.
Después tenemos el plaguicida B que de por sí tiene una eficacia mucho mayor, con un 95% y
mantuvo a raya la población de polillas, aniquilando las polillas después de 3 generaciones.
Cada temporada hay 3 picos de crecimiento de la polilla de la vid, un detalle que no tomamos en
consideración, pero es importante saberlo para aplicar el plaguicida cada vez que llegue la época de
la polilla.
Nuestra hipótesis se cumplió, el plaguicida con mayor eficiencia aniquiló la plaga rápidamente y el
plaguicida con peor eficiencia retrasó el crecimiento, pero de manera insignificante, pués las polillas
de igual manera formaron un “imperio”.
Conclusión
En primer lugar, recalcar la importancia de utilizar un buen plaguicida, de calidad, pues si no, toda la
viña se podría ir abajo y la producción se reduciría, como pasó con la viña que utilizó el plaguicida A.
También recalcar la sensibilidad de la tasa de reproducción de esta especie, sufriendo todos los
cambios, por más insignificante que pueda ser. El crecimiento poblacional puede ser rapidísimo, o
muy lento, todo medible por pequeñas variaciones en “r”.
Cómo última recomendación recalcar que pese a que un plaguicida tenga buena eficiencia hay que
tener en cuenta y ser cuidadoso con sus efectos secundarios, pues pueden llegar a empeorar la
cosecha.
Mantener a raya nuestras plagas es importante para mantener viva nuestra economía y nuestro
puesto como país rico en vinos, la polilla de la vid es uno de los muchos otros ejemplos de factores
biológicos relevantes en nuestra economía y otro de los muchos ejemplos graficables con el mapa
logístico, que es de lo que trato nuestro trabajo como predictores biológicos.
