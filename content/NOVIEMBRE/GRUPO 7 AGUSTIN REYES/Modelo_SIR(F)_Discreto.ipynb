```python
def SIR(S, I, R, F, beta, gamma, alfa, n):
    Sanos = [S]
    Infectados = [I]
    Recuperados = [R]
    Fallecidos = [F]
    for i in range(n):
        Sn = Sanos[-1] - beta * Sanos[-1] * Infectados[-1]
        In = Infectados[-1] + beta * Sanos[-1] * Infectados[-1] - gamma * Infectados[-1] - alfa * Infectados[-1]
        Rn = Recuperados[-1] + gamma * Infectados[-1]
        Fn = Fallecidos[-1] + alfa * Infectados[-1]
        Sanos.append(Sn)
        Infectados.append(In)
        Recuperados.append(Rn)
        Fallecidos.append(Fn)
    return(Sanos, Infectados, Recuperados, Fallecidos)
```

```python
SIR(0.99, 0.01, 0, 0, 0.5, 0.05, 0.01, 20)
```
