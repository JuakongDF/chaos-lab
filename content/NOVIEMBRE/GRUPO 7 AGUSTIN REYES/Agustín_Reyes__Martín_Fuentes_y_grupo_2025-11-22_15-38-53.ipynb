El sistema usado para el trabajo es el sistema SIR discreto. Este sistema creado para analizar el comportamiento de distintas enfermedades está generado a partir de tres ecuaciones:


Donde las variables representan distintos grupos de población a observar durante una epidemia, S es la población susceptible al virus, I es la población infectada por el virus, y R es la población recuperada tras estar infectada.

Además están los otros dos parámetros que se modifican acorde al comportamiento de la epidemia, beta es la tasa de contacto/contagio, es decir, la probabilidad de que exista un contagio por cada contacto. Y gamma representa la tasa de recuperación, cuanta población infectada pasa al grupo R en función del tiempo.

Este sistema omite algunos factores con el fin de simplificar las cosas, no considera el nacimiento de más población, toma a los fallecidos como recuperados, y asume que estos mismos cuentan con inmunidad tras la enfermedad. Debido a esto, la suma de las variables dará siempre un mismo valor de población, debido a que los fallecidos no disminuyen, los parámetros van a tender a estabilizarse, en que toda la población se encontrará en el grupo de recuperados, puesto que los grupos de personas van a ir siempre de S a I, y de I a R, que se toma como inmunidad.

Sin embargo diversas fuentes como Math Insight y ScienceDirect.com han planteado diversos sistemas capaces de representar el verdadero comportamiento de las epidemias, un sistema donde un porcentaje de infectados fallece, otro donde no se toma a los recuperados como grupo inmune, y uno parecido al anterior donde los recuperados ganan inmunidad pero solo por un tiempo, es decir, de R vuelven a S.

El modelo a investigar en el informe es el que considera los fallecidos, esto además modifica la ecuación de infectados y crea una ecuación nueva, dada por la multiplicación entre población de infectados y un valor (alpha) que será un índice de mortalidad, además se tratará a los fallecidos como D.

Conclusión 
El análisis de los gráficos nos permite corroborar la hipótesis inicial, pudimos, al observar gráficos descubrir cómo se comporta la mortalidad de enfermedades al variar diferentes parámetros. Lo que nos permite establecer una serie de medidas importantes a la hora de controlar una enfermedad que se está contagiando rápidamente, ya que al reducir los contactos se reducen contagios, es importante además lograr vacunar a la mayor parte de la población posible, de esta forma aumentamos el grupo que cuenta con inmunidad ante las enfermedades. 

Si bien no tenemos una visualización en un gráfico logramos observar que la secuencia el valor S tiende a 0, el valor I aumenta, llega a un maximo y disminuye rápidamente, luego se desacelera, el valor R, aumenta constantemente, pero disminuye la velocidad con la que aumenta, al igual que el valor D.


```python
def SIR(S, I, R, F, beta, gamma, alfa, n):
    Sanos = [S]
    Infectados = [I]
    Recuperados = [R]
    Fallecidos = [F]
    for i in range(n):
        Sn = Sanos[-1] - beta * Sanos[-1] * Infectados[-1]
        In = Infectados[-1] + beta * Sanos[-1] * Infectados[-1] - gamma * Infectados[-1] - alfa * Infectados[-1]
        Rn = Recuperados[-1] + gamma * Infectados[-1]
        Fn = Fallecidos[-1] + alfa * Infectados[-1]
        Sanos.append(Sn)
        Infectados.append(In)
        Recuperados.append(Rn)
        Fallecidos.append(Fn)
    return(Sanos, Infectados, Recuperados, Fallecidos)
```
