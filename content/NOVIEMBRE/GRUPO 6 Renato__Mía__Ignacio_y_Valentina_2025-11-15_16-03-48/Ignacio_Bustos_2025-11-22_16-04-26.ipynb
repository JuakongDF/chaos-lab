```python
def secuencia(a, u, num):
    x = a
    res = [[],[a]]
    for i in range(num):
        res[0].append(i)
    for i in range(num-1):
        x = generarPasos(u, x)
        res[1].append(x)
    return res
        
```

```python
def mapa_logistico_clasico(r, x):
    """Calcula x_{n+1} usando el mapa cl√°sico."""
    return r * x * (1 - x)

def generar_secuencia_clasica(r, x0, n_pasos):
    """Genera secuencia con el mapa cl√°sico."""
    valores = [[],[x0]]
    for i in range(n_pasos):
        valores[0].append(i)
    for i in range(n_pasos - 1):
        valores[1].append(mapa_logistico_clasico(r, valores[1][i]))
    return valores
    
def cobweb_sequence(start, r, steps):
    xs, ys = [start], []
    x = start
    for i in range(steps):
        y = r * x * (1 - x)
        ys.append(y)
        x = y
        xs.append(x)
    return xs, ys
def lista_atractores(valores, ultimos):
    res =[]
    lista=[]
    for i in range(ultimos):
        lista.append(round(valores[(i+1)*-1], 3))
    for elemento in lista:
        if elemento not in res:
            res.append(elemento)
    return res
def sep_list(valores):
    res = []
    for i in range(len(valores)):
        for j in range(len(valores[i])):
            res.append(valores[i][j])
    return res

def diagrama_bif(ultimos, inicio):
    r = 2
    x = []
    y = []
    for i in range(201):
        r_now = r + (i*0.01)
        sec = generar_secuencia_clasica(r_now, inicio, 100)
        atract = lista_atractores(sec[1], ultimos)
        y.append(atract)
        for i in range(len(atract)):
            x.append(r_now)
    y = sep_list(y)
    return x, y

def diagrama_bifU(ultimos, inicio):
    r = 1
    x = []
    y = []
    for i in range(401):
        r_now = r + (i*0.0025)
        sec = secuencia(inicio, r_now, 100)
        atract = lista_atractores(sec[1], ultimos)
        y.append(atract)
        for i in range(len(atract)):
            x.append(r_now)
    y = sep_list(y)
    return x, y
```

```python
#definicion de variables
mu = 2
start = 0.9
steps = 50
r = 2
x_range = [0, 100]
y_range = [0, 1]
y_dtick = 0.1
telara√±a = False
opcion = ["Caos determinista", "Modelo log√≠stico", "Ambos", "Diagrama de telara√±a", "Diagrama de bifurcacion", "Diagrama de bifurcacion determinista"]
def update(mu=2.0, start=0.9, steps=50, r=2.0, opcion=opcion):
    if opcion == "Caos determinista":
        fig = make_subplots(rows=1, cols=1,subplot_titles=[f'Caos determinista(mu={mu:.2f})'])
        xs, ys = secuencia(start, mu, steps)
        det = go.Scatter(x=xs, y=ys, mode="lines+markers", name="modelo caos determinista")
        fig.add_trace(det, 1, 1)
    elif opcion == "Modelo log√≠stico":
        fig = make_subplots(rows=1, cols=1,subplot_titles=[f'Mapa log√≠stico(r={r:.2f})'])
        xl, yl = generar_secuencia_clasica(r, start, steps)
        log = go.Scatter(x=xl, y=yl, mode="lines+markers", name="modelo log√≠stico")
        fig.add_trace(log, 1, 1)
    elif opcion == "Ambos":
        fig = make_subplots(rows=1, cols=1,subplot_titles=[f'caos determinista y mapa log√≠stico(mu={mu:.2f} r={r:.2f})'])
        xs, ys = secuencia(start, mu, steps)
        xl, yl = generar_secuencia_clasica(r, start, steps)
        det = go.Scatter(x=xs, y=ys, mode="lines+markers", name="modelo caos determinista")
        log = go.Scatter(x=xl, y=yl, mode="lines+markers", name="modelo log√≠stico")
        fig.add_trace(det, 1, 1)
        fig.add_trace(log, 1, 1)
    elif opcion == "Diagrama de telara√±a":
        fig = make_subplots(rows=1, cols=1,
                            subplot_titles=[f'profe pongame un 7 (r={r:.2f})'])
    
        # L√≠nea identidad
        diag_x = [0, 1]
        diag_y = [0, 1]
    
        # Mapa log√≠stico
        map_x = [i/100 for i in range(101)]
        map_y = [r * xi * (1 - xi) for xi in map_x]
    
        fig.add_trace(go.Scatter(
            x=diag_x, y=diag_y,
            mode='lines', line=dict(color='black', dash='dash'),
            name='y = x'
        ), row=1, col=1)
    
        fig.add_trace(go.Scatter(
            x=map_x, y=map_y,
            mode='lines', line=dict(color='green'),
            name='y = r*x*(1-x)'
        ), row=1, col=1)
    
        # Secuencia para el diagrama
        cobweb_xs, cobweb_ys = cobweb_sequence(start, r, steps)
    
        for i in range(steps):
            # L√≠neas verticales
            fig.add_trace(go.Scatter(
                x=[cobweb_xs[i], cobweb_xs[i]],
                y=[cobweb_xs[i], cobweb_ys[i]],
                mode='lines', line=dict(color='blue', width=1)
            ), row=1, col=1)
    
            # L√≠neas horizontales
            fig.add_trace(go.Scatter(
                x=[cobweb_xs[i], cobweb_ys[i]],
                y=[cobweb_ys[i], cobweb_ys[i]],
                mode='lines', line=dict(color='red', width=1)
            ), row=1, col=1)
    elif opcion == "Diagrama de bifurcacion":
        fig = make_subplots(rows=1, cols=1,subplot_titles=[f'Diagrama de bifurcacion(inicio={start:.2f})'])
        xb, yb = diagrama_bif(20, start)
        bif = go.Scatter(x=xb, y=yb, mode="markers", name="Diagrama de bifurcacion")
        fig.add_trace(bif, 1, 1)
    elif opcion == "Diagrama de bifurcacion determinista":
        fig = make_subplots(rows=1, cols=1,subplot_titles=[f'Diagrama de bifurcacion determinista(inicio={start:.2f})'])
        xb, yb = diagrama_bifU(20, start)
        bif = go.Scatter(x=xb, y=yb, mode="markers", name="Diagrama de bifurcacion")
        fig.add_trace(bif, 1, 1)
    if opcion == "Diagrama de telara√±a":
        fig.update_xaxes(range=[0,1], row=1, col=1)
        fig.update_yaxes(range=y_range, row=1, col=1, dtick=y_dtick)
    elif opcion == "Diagrama de bifurcacion":
        fig.update_xaxes(range=[2,4], row=1, col=1)
        fig.update_yaxes(range=y_range, row=1, col=1, dtick=y_dtick)
    elif opcion == "Diagrama de bifurcacion determinista":
        fig.update_xaxes(range=[1,2], row=1, col=1)
        fig.update_yaxes(range=y_range, row=1, col=1, dtick=y_dtick)
    else:
        fig.update_xaxes(range=x_range, row=1, col=1)
        fig.update_yaxes(range=y_range, row=1, col=1, dtick=y_dtick)
    fig.update_layout(title="an√°lisis del modelo de caos determinista y mapa log√≠stico", width=900, height=900, showlegend=False)
    display(HTML(to_html(fig, include_plotlyjs="cdn", full_html=False)))

interact(update, mu=(0, 2, 0.1), start=(0, 1, 0.1), steps=(1, 100, 1), r=(0, 4, 0.1))
```
